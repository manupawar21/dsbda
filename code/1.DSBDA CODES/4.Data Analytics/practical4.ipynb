```python
X = df[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','BLACK','LSTAT']]
Y = df['MEDV']
```

```python
#Trainning Linear Regression Model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,random_state=0)
```

```python
print("Train DataSet Size = ", "X : ", X_train.shape, "Y : ", Y_train.shape)
print("Test DataSet Size = ", "X : ", X_test.shape, "Y : ", Y_test.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
prediction = lr.predict(X_test)
```

```python
plt.figure(figsize = (6,6))
plt.scatter(Y_test, prediction)
plt.xlabel("Y test")
plt.ylabel("Predicted Y")
plt.title("Test vs Prediction")
```
